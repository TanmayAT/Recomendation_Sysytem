```python
 #Calculate the frequency of each payment mode
payment_mode_counts = df['Payment Mode'].value_counts()

# Extract values and labels for the pie chart
values = payment_mode_counts.values
labels = payment_mode_counts.index

# Create a pie chart
plt.pie(values, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Payment Mode Distribution')
plt.axis('equal')
```

```python
 #Calculate the frequency of each payment mode
quantity_mode_counts = df["Quantity of Items"].value_counts()

# Extract values and labels for the pie chart
values = quantity_mode_counts.values
labels = quantity_mode_counts.index

# Create a pie chart
plt.pie(values, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Quantity of Items')
plt.axis('equal')
```

```python
 #Calculate the frequency of each payment mode
rating_mode_counts = df["Customer Rating-Food"].value_counts()

# Extract values and labels for the pie chart
values = rating_mode_counts.values
labels = rating_mode_counts.index

# Create a pie chart
plt.pie(values, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('rating of Items')
plt.axis('equal')
```
