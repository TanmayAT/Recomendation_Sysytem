# budget Based Recommendation system 

```python

# Create a pivot table to summarize sales by date and product
pt = pd.pivot_table(df, values= 'Budget', index='Cuisines', columns='User ID')

pt.fillna(0)
```

```python
for index, row in pt.iterrows():
    # Loop through each column in the row
    for column in pt.columns:
        # Check if the value is NaN
        if pd.isna(row[column]):
            # Replace NaN with a specific value (e.g., 0)
            pt.at[index, column] = 0
```

```python
pt.isnull().sum()
```

```python
from sklearn.metrics.pairwise import cosine_similarity

cosine_sim = cosine_similarity(pt) 
```

```python
def recommend(Cuisines) :
    
    index = np.where(pt.index == Cuisines)[0][0]
    
    distances   = sorted(list(enumerate(cosine_sim[0])),key = lambda  x:x[1], reverse = True )
    
    for i in distances :
        print(i)
    
```

```python
recommend('Chinese')
```
